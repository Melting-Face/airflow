```python
import pandas as pd

from os import listdir

csvs = listdir('csv')
for csv in csvs:
    df = pd.read_csv(f'csv/{csv}')
    print(csv, df.shape[0])
    print(df.columns)
    columns = [column for column in df.columns if 'HS ' in column]
    output_df = pd.DataFrame()
    df = df.loc[(df[columns[0]].notnull()) & (df[columns[1]].notnull())]
    df = df.astype(str)
    output_df['code_from'] = df[columns[0]].apply(str).str.pad(width=6, side='left', fillchar='0')
    output_df['code_to'] = df[columns[1]].str.pad(width=6, side='left', fillchar='0')
    output_df['relationship'] = df['Relationship']

    columns = [column.replace('HS ', '') for column in columns]
    output_df['code_from'] = output_df['code_from'].str.replace('.0', '')
    output_df['code_to'] = output_df['code_to'].str.replace('.0', '')
    output_df['classification_from'] = columns[0]
    output_df['classification_to'] = columns[1]
    if 'ex' in df.columns:
        output_df['ex'] = df['ex']
    else:
        output_df['ex'] = ''
    # output_df['code_from'] = output_df['code_from'].str.split('.')[0]
    output_df = output_df.fillna('')
    print(columns[0], columns[1], df.shape[0])
    print(output_df.dtypes)
    print(output_df.tail())
    
    output_df.to_parquet(f'parquet/{columns[0]}to{columns[1]}.parquet')
```
